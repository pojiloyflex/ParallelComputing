```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
```

```python
time = pd.DataFrame({
    "50dots": [0.082030, 0.109086, 0.109260, 0.145743],
    "1000dots": [15.998799, 12.205623, 9.325433, 6.962459],
    "threads": [1, 2, 4, 8]
})

boost = pd.DataFrame({
    "50dots": [1, 0.7519755056, 0.7507779608, 0.5628400678],
    "1000dots": [1, 1.3107728299, 1.7156092377, 2.2978661706],
    "threads": [1, 2, 4, 8]
})

effectiveness = pd.DataFrame({
    "50dots": [1, 0.3759877528, 0.1876944902, 0.0703550085],
    "1000dots": [1, 0.6553864149, 0.4289023094, 0.2872332713],
    "threads": [1, 2, 4, 8]
})
```

```python
display(time)
display(boost)
display(effectiveness)
```

```python
def get_line_plot(df: pd.DataFrame, x: str, y: str, title: str = None):
    fig = px.line(df, x=x, y=y, markers=True)
    fig.update_layout(
        title=title,
        width=700,
    )

    return fig
```

```python
(
    get_line_plot(time, "threads", "50dots", "50 dots time comparison")
    .update_layout(xaxis_title="Number of threads", yaxis_title="Walltime")
    .update_xaxes(type='category')
).show()

(
    get_line_plot(time, "threads", "1000dots", "1000 dots time comparison")
    .update_layout(xaxis_title="Number of threads", yaxis_title="Walltime")
    .update_xaxes(type='category')
).show()
```

```python
(
    get_line_plot(boost, "threads", "50dots", "50 dots boost")
    .update_layout(xaxis_title="Number of threads", yaxis_title="Boost value")
    .update_xaxes(type='category')
).show()

(
    get_line_plot(boost, "threads", "1000dots", "1000 dots boost")
    .update_layout(xaxis_title="Number of threads", yaxis_title="Boost value")
    .update_xaxes(type='category')
).show()
```

```python
(
    get_line_plot(effectiveness, "threads", "50dots", "50 dots effectiveness")
    .update_layout(xaxis_title="Number of threads", yaxis_title="Effectiveness")
    .update_xaxes(type='category')
).show()

(
    get_line_plot(effectiveness, "threads", "1000dots", "1000 dots effectiveness")
    .update_layout(xaxis_title="Number of threads", yaxis_title="Effectiveness")
    .update_xaxes(type='category')
).show()
```
